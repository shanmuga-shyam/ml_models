```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel(r"C:\Users\Shanmuga Shyam. B\OneDrive\Desktop\ml-learn\linear_regression\linear_example.xlsx")
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]],df.price)
```

```python
plt.scatter(df.area,df.price,color="red",marker="+")
plt.xlabel("area",fontsize=20)
plt.ylabel("price",fontsize=20)
plt.plot(df.area,reg.predict(df[["area"]]),color="teal")
```

```python
import pickle
with open("model_pickel","wb") as f:
    pickle.dump(reg,f)
with open("model_pickel","rb") as f:
    mp = pickle.load(f)
```

```python
mp.predict(pd.DataFrame({"area":[2000]}))
```
