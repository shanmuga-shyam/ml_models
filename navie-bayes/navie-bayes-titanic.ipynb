```python
import os
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Shanmuga Shyam. B\OneDrive\Desktop\ml-learn\navie-bayes\titanic.csv")
df.head()
```

```python
df.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis = True, inplace=True)
df.head()
```

```python
target = df.Survived
inputs = df.drop("Survived",axis = "columns")
inputs.head()
```

```python
dummies = pd.get_dummies(inputs.Sex, dtype = int)
dummies.head(3)
```

```python
inputs = pd.concat([inputs, dummies], axis="columns")
inputs.head(3)
```

```python
inputs = inputs.drop("Sex",axis="columns")
inputs.head()
```

```python
 inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head(5)
```

```python
 inputs.columns[inputs.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs,target,test_size=0.7)
```

```python
print(len(x_train))
print(len(x_test))
print(len(inputs))
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
